```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
```

```python
load_dotenv()
```

```python
model = ChatGoogleGenerativeAI(model='gemini-2.5-flash')
```

```python
chat_history =[SystemMessage(content='You are a personalize assistent how answer based on the question and also on prevoius questin.')]
```

```python
chat_history
```

```python
while True:
    user_input = input('You : ')
    query = f'previous history : {chat_history} asnd current question : {user_input}'
    if(user_input == 'exit'):
        break
    result = model.invoke(query)
    user = f'User : {user_input}'
    response = f'Response : {result.content}'
    chat_history.append(HumanMessage(content = user_input))
    chat_history.append(AIMessage(content=result.content))
    print(user)
    print(response)
```
