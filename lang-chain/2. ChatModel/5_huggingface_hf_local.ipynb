```python
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline
from transformers import pipeline
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
import re

def get_response(response):
    text = response.content  

    user_match = re.search(r"<\|user\|\>\n(.*?)</s>", text, re.DOTALL)
    assistant_match = re.search(r"<\|assistant\|\>\n(.*)", text, re.DOTALL)

    user_input = user_match.group(1).strip() if user_match else None
    assistant_output = assistant_match.group(1).strip() if assistant_match else None

    return assistant_output
```

```python
pipe = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0"
)
```

```python

llm = HuggingFacePipeline.from_model_id(
    model_id='TinyLlama/TinyLlama-1.1B-Chat-v1.0',
    task='text-generation',
)
```

```python
model = ChatHuggingFace(llm=llm)
```

```python
while True:
    user_input = input('You : ')
    if(user_input == 'exit'):
        break
    result = model.invoke(user_input)
    print("Response : ", get_response(result))
```
